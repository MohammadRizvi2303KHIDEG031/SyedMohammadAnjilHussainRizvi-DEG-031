# taking a parquet file and manipulating its earings colummn to produce a new parquet file

```python
import pandas as pd 
import numpy as np 
```

```python
df_1 = pd.read_parquet('/home/syedmohammadanjilhussainrizvi/Desktop/data_engineering_bootcamp_2303/tasks/5_data_pipelines/day_4_data_lake/data/output_data/employee_earnings/earnings_date=2022-02-10/employee_earnings.parquet') 
```
