```python
import pandas as pd

def Label_encoding(daata):
    listt = []
    for j,i in enumerate(daata.columns):
        
        if daata.iloc[:,j].dtype == 'O':
            d =daata[i].drop_duplicates().reset_index(drop=True)
            g =dict(zip([i for i in d],[i for i in d.index]))
            
            for j in daata[i]:
                listt.append(g[j])
                
            daata[f"Label_encoded {i}"] = listt
            listt = []
```

```python
dataframe = pd.DataFrame({"Names":["Taha","Ubaid","Ali","Huzefa","Taha","Talha","Talha","Huzefa","Talha"],
             "Department":["CS","CS","SE","SE","CS","EE","EE","SE","EE"],
             "col":[1,2,3,4,5,6,7,8,9]})

Label_encoding(dataframe)
dataframe
```
